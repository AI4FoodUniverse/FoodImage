### convert dng to jpg

```python
import os
from PIL import Image
import shutil


input_folder = "rawdata/batch4"

output_folder = "rawdata/batch4crop"


if not os.path.exists(output_folder):
    os.makedirs(output_folder)

for file_name in os.listdir(input_folder):

    if file_name.endswith(".dng"):
        dng_file_path = os.path.join(input_folder, file_name)
        
        dng_image = Image.open(dng_file_path)
        

        jpeg_file_path = os.path.join(output_folder, file_name[:-4] + ".jpg")
        
        dng_image.save(jpeg_file_path)

        print(f"Converted {dng_file_path} to {jpeg_file_path}")

    elif file_name.endswith(".jpg"):
        jpeg_file_path = os.path.join(input_folder, file_name)
        output_jpeg_file_path = os.path.join(output_folder, file_name)
        shutil.copy(jpeg_file_path, output_jpeg_file_path)
        
        print(f"Copied {jpeg_file_path} to {output_jpeg_file_path}")

print("Conversion completed.")

```

### resize image

```python
from PIL import Image
import os

def batch_resize(input_folder, output_folder):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    
    for filename in os.listdir(input_folder):
        if filename.endswith('.jpg'):
            image_path = os.path.join(input_folder, filename)
            img = Image.open(image_path)
            img_resized = img.resize((256, 256))
            img_resized.save(os.path.join(output_folder, filename))

# input_folder_base = 'rawdataset/Day{}crop'
# output_folder_base = 'rawdataset/Day{}resize'
# for day in range(1, 9):
#     if day == 7: 
#         continue
#     input_folder = input_folder_base.format(day)
#     output_folder = output_folder_base.format(day)
#     batch_resize(input_folder, output_folder)
batch_resize('rawdata/batch4cropped', 'rawdata/batch4resized')
```
